```python
from ipythonblocks import ImageGrid
```

```python
grid = ImageGrid(10, 10, fill=(124, 124, 124), block_size=15)
grid
```

```python
for pixel in grid:
    if pixel.x < 5:
        pixel.green = 255
    elif pixel.x > 5 and pixel.y >= 5:
        pixel.blue = 255
grid
```

```python
grid[:3, :3] = (255, 124, 124)
grid
```

```python
grid[::4, ::4] = (0, 0, 0)
grid
```
